```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_ex = pd.read_csv('./exemp.csv')
df_nonex = pd.read_csv('./nonexemp.csv')
```

```python
sal_ex = df_ex.groupby('salary').size()
sal_nonex = df_nonex.groupby('salary').size()

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 0.75, 0.75], aspect=1)
ax1.pie(sal_ex.values,labels = sal_ex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Existing Emp salary dist')

ax2 = fig.add_axes([1, 0, 0.75, 0.75], aspect=1)
ax2.pie(sal_nonex.values,labels = sal_nonex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Non Existing Emp salary dist')
```

```python
prom_ex = df_ex.groupby('promotion_last_5years').size()
prom_nonex = df_nonex.groupby('promotion_last_5years').size()

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 0.75, 0.75], aspect=1)
ax1.pie(prom_ex.values,labels = prom_ex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Existing Emp promotions dist')

ax2 = fig.add_axes([1, 0, 0.75, 0.75], aspect=1)
ax2.pie(prom_nonex.values,labels = prom_nonex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Non Existing Emp promotions dist')
```

```python
dept_ex = df_ex.groupby('dept').size()
dept_nonex = df_nonex.groupby('dept').size()

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 0.75, 0.75], aspect=1)
ax1.pie(dept_ex.values,labels = dept_ex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Existing Emp dept dist')

ax2 = fig.add_axes([1, 0, 0.75, 0.75], aspect=1)
ax2.pie(dept_nonex.values,labels = dept_nonex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Non Existing Emp dept dist')
```

```python
dept_ex = df_ex.groupby('number_project').size()
dept_nonex = df_nonex.groupby('number_project').size()

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 0.75, 0.75], aspect=1)
ax1.pie(dept_ex.values,labels = dept_ex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Existing Emp dept dist')

ax2 = fig.add_axes([1, 0, 0.75, 0.75], aspect=1)
ax2.pie(dept_nonex.values,labels = dept_nonex.index,startangle = 90, autopct='%1.1f%%');
plt.title('Non Existing Emp dept dist')
```
